### PDF ingestion pipeline

```python
from collections import defaultdict

def load_and_combine_pdfs(path):
    """"""
    # Read and parse directory path
    reader = SimpleDirectoryReader(input_dir=path)
    documents = reader.load_data()
    # Group docs by their source file
    file_pages = defaultdict(list)
    for doc in documents:
        file = doc.metadata["file_name"]
        file_pages[file].append(doc.text)

    joined_files = {key: "\n".join(value) for key, value in file_pages.items()}
    return joined_files
```

```python
med_dict = load_and_combine_pdfs(path="./PDFs/Medline/")

for filename, text in med_dict.items():
    print(f"{filename}: {len(text)} chars")

print(med_dict['a1c.pdf'])
```

```python
# Cleaning function to split on "References\n" & keep all before it for Medline docs 

def clean_medline(combined_str):
    substring = "References\n"
    try:
        substring_idx = combined_str.index(substring)
        return combined_str[:substring_idx]
    except ValueError:
        print(f"The substring '{substring}' was not found and caused a ValueError.")
        return combined_str
```

```python
cleaned_str = clean_medline(med_dict['a1c.pdf'])
print(cleaned_str.find("References\n"))
```

```python
# Cleaning function for Mayo pdfs

import re

def clean_mayo(combined_str):
    fees_substring = "Fees & Codes\n"
    separator = "___________________________________________________\n___________________________________________________\n"
    try:
        fees_idx = combined_str.index(fees_substring)
        trimmed = combined_str[:fees_idx]
        chunks = trimmed.split(separator)
        cleaned_chunks = [chunks[0]]
        for chunk in chunks[1:]:
            chunk = chunk.split('\n', 1)[1]           # drop "Document generated..." line
            chunk = re.sub(r'^Page \d+\s+of \d+\s*', '', chunk)  # drop "Page X  of N" prefix
            cleaned_chunks.append(chunk)
        return '\n'.join(cleaned_chunks)
    except ValueError:
        print(f"The substring '{fees_substring}' was not found.")
        return combined_str
 
```

### Apply appropriate functions to both directories

```python
# Orchestrator function to load, combine, & clean Medline & Mayo pdfs

def load_combine_clean_pdfs(med_path, mayo_path):
    """"""
    # Load & combine pdfs from each dir
    medline_dict = load_and_combine_pdfs(med_path)
    mayo_dict = load_and_combine_pdfs(mayo_path)

    #
    medline_final = []
    for filename, text in medline_dict.items():
        cleaned_str = clean_medline(text)
        source = "medline"
        test_name = filename.replace('.pdf', '')
        add_med_dict = {"text": cleaned_str, "source": source, "test_name": test_name}
        medline_final.append(add_med_dict)
    
    #
    mayo_final = []
    for filename, text in mayo_dict.items():
        cleaned_str = clean_mayo(text)
        source = "mayo"
        test_name = filename.replace('.pdf', '')
        add_mayo_dict = {"text": cleaned_str, "source": source, "test_name": test_name}
        mayo_final.append(add_mayo_dict)

    return medline_final, mayo_final
        

MED_PATH = "./PDFs/Medline/"
MAYO_PATH = "./PDFs/Mayo/"

medline_final, mayo_final = load_combine_clean_pdfs(med_path=MED_PATH, mayo_path=MAYO_PATH)
```

### Reranking Experiment

Yes, re-ranking is the next step.

Right now your retrieval pipeline is single-stage:

1. Embed the query
2. Find the top-k chunks by cosine similarity in ChromaDB
3. Pass those chunks directly to the LLM

The problem is cosine similarity is a blunt instrument — it measures how similar two vectors are in embedding space, but embeddings compress meaning. Two chunks can land close together in embedding space even if one is much more relevant to the specific query than the other. That's why "HbA1c tube type" pulls a Medline chunk when the Mayo chunk exists.

Re-ranking adds a second stage:

1. Embed the query → retrieve top-k chunks by cosine similarity (same as before, but you cast a wider net — e.g., top 10 instead of top 2)
2. Pass each of those 10 candidate chunks through a cross-encoder model alongside the query
3. The cross-encoder reads both the query and the chunk together and outputs a relevance score
4. Re-sort by that score and keep only the top 2 to pass to the LLM

Why is a cross-encoder better than cosine similarity?

Embedding models encode the query and chunk separately — they never see each other. A cross-encoder reads them together, which lets it reason about how well this specific chunk actually answers this specific question. That's a fundamentally richer comparison.

The tradeoff is speed — running a cross-encoder over 10 candidates is slower than a single cosine similarity lookup. That's why you do it as a second pass over a small candidate set, not over the entire collection.

In LlamaIndex, this is implemented as a postprocessor that sits between retrieval and generation.

```python
from llama_index.postprocessor.sbert_rerank import SentenceTransformerRerank
logging.getLogger("httpx").setLevel(logging.WARNING)

reranker = SentenceTransformerRerank(
    model="cross-encoder/ms-marco-MiniLM-L-6-v2",
    top_n=2
)

query_engine = index.as_query_engine(
    similarity_top_k=6,
    node_postprocessors=[reranker],
    text_qa_template=qa_template
)

scores = []

# Judge LLM
judge_llm = OpenAI(model="gpt-4o")
evaluator = CorrectnessEvaluator(llm=judge_llm)

for pair in combined_evals:
    question = pair['question']
    expected = pair['expected_answer']

    response = query_engine.query(question)

    result = evaluator.evaluate(
        query=question,
        response=str(response),
        reference=expected
    )
    scores.append(result.score)
    print(f"Question: {question}")
    print(f"Expected: {expected}")
    print(f"Response: {response}")
    print(f"Score: {result.score} / 5.0")
    print(f"Feedback: {result.feedback}")
    print("="*60)

print(f"\nAverage Score: {round(np.mean(scores), 2)} / 5.0")
```

### Agent Layer

```python
from llama_index.core.llms import ChatMessage

# Intent classifier — gates out-of-scope, emergency, and diagnosis queries
CLASSIFIER_SYSTEM_PROMPT = """You are a query classifier for a medical laboratory test information assistant.
Classify the user's query into exactly one of these categories:

- in_scope: The query asks about laboratory tests — what they measure, why they are ordered, how to prepare, what results mean, reference ranges, specimen handling, tube types, or collection procedures.
- out_of_scope: The query is not related to laboratory tests.
- medical_emergency: The query describes an urgent medical situation, severe symptoms, or asks if the user is in danger.
- diagnosis_request: The query asks for a personal medical diagnosis or whether the user has a specific condition.

Respond with ONLY the category label. No explanation. No punctuation. Just the label."""

def classify_query(query: str) -> str:
    messages = [
        ChatMessage(role="system", content=CLASSIFIER_SYSTEM_PROMPT),
        ChatMessage(role="user", content=query)
    ]
    response = Settings.llm.chat(messages)
    return response.message.content.strip().lower()


# Source classifier — routes in-scope queries to the correct engine
SOURCE_CLASSIFIER_PROMPT = """You are a routing classifier for a medical laboratory assistant with two source databases:
- patient_education: MedlinePlus articles written for patients covering explanations, purposes, and preparation
- lab_procedures: Mayo Clinic lab handbook with specimen collection protocols and reference values

Classify the query into exactly one of these categories:

- patient_education: Use when the question asks WHY a test is ordered, WHAT a test is or does conceptually, WHAT results mean for a patient, HOW TO PREPARE for a test, or compares tests at a conceptual level.
  Examples: "Why do I need a CBC?", "What does a high WBC indicate?", "Do I need to fast?", "What is the difference between BMP and CMP?"

- lab_procedures: Use when the question asks for specific collection requirements (tube type, volume, handling), specimen stability windows, specific numeric reference ranges by age or sex, or clinical diagnostic thresholds.
  Examples: "What tube type is required?", "What is the hemoglobin reference range for males 6-8 years?", "How long is the specimen stable?", "What HbA1c threshold does the ADA use?"

Respond with ONLY the category label. No explanation. No punctuation. Just the label."""

def classify_source(query: str) -> str:
    messages = [
        ChatMessage(role="system", content=SOURCE_CLASSIFIER_PROMPT),
        ChatMessage(role="user", content=query)
    ]
    response = Settings.llm.chat(messages)
    return response.message.content.strip().lower()
```

```python
from llama_index.core.vector_stores import MetadataFilters, ExactMatchFilter

# Create source-specific query engines with metadata filters
medline_filters = MetadataFilters(filters=[ExactMatchFilter(key='source', value='medline')])
mayo_filters = MetadataFilters(filters=[ExactMatchFilter(key='source', value='mayo')])

# Medline engine
medline_engine = index.as_query_engine(
    similarity_top_k=6,
    node_postprocessors=[reranker],
    text_qa_template=qa_template,
    filters=medline_filters
)
# Mayo engine
mayo_engine = index.as_query_engine(
    similarity_top_k=6,
    node_postprocessors=[reranker],
    text_qa_template=qa_template,
    filters=mayo_filters
)
```

```python
# Router function that:
    # Determines category of query & handles out of scope queries
    # Routes in scope queries to classify_source to determine source query should use
    # Matches source to correct query engine (Medline or Mayo)
    # If query is out of scope, then respond using CANNED_RESPONSES w/ out of scope fallback

CANNED_RESPONSES = {
    "out_of_scope": "I can only answer questions about laboratory tests. Please ask about a specific lab test.",
    "medical_emergency": "If you are experiencing a medical emergency, call 911 immediately. This assistant cannot help with urgent medical situations.",
    "diagnosis_request": "I'm not able to provide a personal medical diagnosis. Please consult your healthcare provider to discuss your specific results."
}
# Note: update out_of_scope to include expanded catalog later

def route_query(query: str):
    category = classify_query(query)
    print(f"[intent classifier] {category}")
    if category == "in_scope":
        source = classify_source(query)
        print(f"[source classifier] {source}")
        if source == "lab_procedures":
            return mayo_engine.query(query)
        else:
            return medline_engine.query(query)
    else:
        return CANNED_RESPONSES.get(category, CANNED_RESPONSES['out_of_scope'])
```

```python
# Evaluate route_query using full CorrectnessEvaluator loop for an updated score
# Should see a jump for Q11 from 2.0

scores = []

# Judge LLM
judge_llm = OpenAI(model="gpt-4o")
evaluator = CorrectnessEvaluator(llm=judge_llm)

for pair in combined_evals:
    question = pair['question']
    expected = pair['expected_answer']

    response = route_query(question)

    result = evaluator.evaluate(
        query=question,
        response=str(response),
        reference=expected
    )
    scores.append(result.score)
    print(f"Question: {question}")
    print(f"Expected: {expected}")
    print(f"Response: {response}")
    print(f"Score: {result.score} / 5.0")
    print(f"Feedback: {result.feedback}")
    print("="*60)

print(f"\nAverage Score: {round(np.mean(scores), 2)} / 5.0")
```

```python
# Full CorrectnessEvaluator on route_query w/ expanded eval set
from llama_index.core.evaluation import FaithfulnessEvaluator

scores = []
faith = []

# Judge LLM
judge_llm = OpenAI(model="gpt-4o")
evaluator = CorrectnessEvaluator(llm=judge_llm)
faithfulness_evaluator = FaithfulnessEvaluator(llm=judge_llm)

for pair in new_combined_evals:
    question = pair['question']
    expected = pair['expected_answer']

    response = route_query(question)

    # Faithfulness Evaluator - if query is not in_scope, append None
    if not isinstance(response, str):
        faith_result = faithfulness_evaluator.evaluate_response(
            query=question,
            response=response
        )
        faith.append(faith_result.score)
    else:
        faith.append(None) # canned response - no source nodes to evaluate

    # Correctness Evaluator
    result = evaluator.evaluate(
        query=question,
        response=str(response),
        reference=expected
    )
    scores.append(result.score)
    print(f"Question: {question}")
    print(f"Expected: {expected}")
    print(f"Response: {response}")
    print(f"Score: {result.score} / 5.0")
    print(f"Feedback: {result.feedback}")
    print("="*60)

print(f"\nAverage Score: {round(np.mean(scores), 2)} / 5.0")
print(f"\nFaith Score: {np.sum([s for s in faith if s is not None])} / 36")
```
